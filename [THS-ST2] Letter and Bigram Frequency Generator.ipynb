```python
import itertools
import nltk
from collections import Counter
```

```python
# replace content with actual corpus to be used
content = "The quick brown fox jumped over the lazy dog."


text_file = open("corpora/Taglish7030.txt", "r")
 
#read whole file to a string
content = text_file.read().replace('“','"').replace('”','"').replace("‘", "'").replace("’","'")

 
#close file
text_file.close()
```

```python
# Define the characters to be included in the list of bigrams
characters = ["q", "w", "e", "r", "t", "y", "u", "i", "o", "p", "[", "]", "\\", "{", "}", "|",
        "a", "s", "d", "e", "f", "g", "h", "j", "k", "l", ";", "'", ":","\"",
        "z", "x", "c","v", "b", "n", "m", "<", ">", "/", ",", ".", "?",
        "`", "1", "2", "3", "4", "5", "6", "7", "8", "9", "0", "-", "=",
        "~", "!", "@", "#", "$", "%", "^", "&", "*", "(", ")", "_", "+", " ",
        "Q", "W", "E", "R", "T", "Y", "U", "I", "O", "P", 
        "A", "S", "D", "F", "G", "H", "I", "J", "K", "L",
        "Z", "X", "C", "V", "B", "N", "M"]
              
# create all the possible combinations
perms = (list(itertools.permutations("".join(characters), 2)))

# initialize dictionary
bigrams = {}
letters = {}

# assign 0.0 as the initial value for every pair
for perm in perms:
    bigrams["{}{}".format(perm[0], perm[1])] = 0.0  

# assign 0.0 as the initial value for every letter
for ch in characters:
    letters[ch] = 0.0
```

```python
# convert string to list
chars = list(content)

# create a counter for the bigrams
cntr_bigram = Counter(zip(chars, chars[1:]))
total_bigram = sum(cntr_bigram.values())

# create a counter for the letters
cntr_chars = Counter(chars)
total_chars = sum(cntr_chars.values())

# create a normalized frequency count for the bigrams of the corpus 
temp_bigram = {''.join(k): int(v) / total_bigram for k,v in cntr_bigram.most_common()}

# create a normalized frequency count for the letters of the corpus 
temp_chars = {''.join(k): int(v) / total_chars for k,v in cntr_chars.most_common()}


# add the produced keys to the bigrams dictionary
for key in temp_bigram:
    if key in bigrams:
        bigrams[key] = temp_bigram[key]
    
# add the produced keys to the letters dictionary
for key in temp_chars:
    if key in letters:
        letters[key] = temp_chars[key]
```

```python
# create file for bigrams count
with open('frequency_bigrams_Taglish7030_QWERTY.txt', 'w') as f:
    for key in bigrams:
        string = "{} {} {}".format(key[0].replace(" ", "space"), key[1].replace(" ", "space"), bigrams[key])
        f.write(f"{string}\n")
```

```python
with open('frequency_letters_Taglish7030_QWERTY.txt', 'w') as f:
    f.write(f"letter frequency\n")
    for key in letters:
        string = "{} {}".format(key.replace(" ", "space"), letters[key])
        f.write(f"{string}\n")
```
